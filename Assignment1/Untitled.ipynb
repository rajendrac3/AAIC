# Exploratory data analysis for haberman dataset

### Importing packages and loading dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

1. Age of patient at time of operation (numerical)
2. Patient's year of operation (year - 1900, numerical)
3. Number of positive axillary nodes detected (numerical)
4. Survival status (class attribute)
        1 = the patient survived 5 years or longer
        2 = the patient died within 5 year

```python
haberman = pd.read_csv('haberman.csv')
print(haberman.shape)  #how many data-points and features
print(haberman.head()) 
```

```python
haberman.columns #column-names in dataset
```

```python
haberman['Survival status'].value_counts() #No.of count of each classes
```

* This is an imbalanced dataset as the number of data for both classes is not equally divided

# 2D-Scatter plot

```python
haberman.plot(kind='scatter', x='Age', y='Year operation')
plt.grid()
plt.show()
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(haberman, hue='Survival status', height=4) \
    .map(plt.scatter, 'Year operation', 'Age') \
    .add_legend()
plt.show()

# It shows as the year of operation increases people survive more. 
```
